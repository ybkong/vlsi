```python
import numpy as np
import pandas as pd
```

```python
# v1,v2 - nor2 nand3 stage4, 16 segment, H=1.92, f = 3.6
# decoder delay 206ps 총delay 1.40ns
Cin = 5 * 1e-15 * 32 
H = 1.92
f = 3.6
Cout = 307.2 * 1e-15
TR_C = 1.36 * 1e-15 / 1e-6

C = [0,  0, 0, 0, Cout]
G = [1, 1.7, 1.6, 1]
W_nmos = [0, 0, 0, 0]
W_pmos = [0, 0, 0, 0]

for i in range(3,-1,-1):
    out = C[i + 1]
    g = G[i]
    C[i] = g*out/f


for i in range(len(G)):
    g = G[i]
    if g==1:
        # nmos W pmos 7/3 W
        # 10/3 * W * TR_C = C
        W_nmos[i] = C[i] / (TR_C * (10/3))
        W_pmos[i] = W_nmos[i] * 7/3
    elif g==1.7: #nor2
        # nmos W pmos 14/3 W
        # 17/3 * W * TR_C = C
        W_nmos[i] = C[i] / (TR_C * (17/3))
        W_pmos[i] = W_nmos[i] * 14/3
    elif g==1.6: #nand3
        # nmos 3W pmos 7/3 W
        # 16/3 * W * TR_C = C        
        W = C[i] / (TR_C * (16/3))
        W_nmos[i] =  W * 3
        W_pmos[i] = W * 7/3 
    else:
        print('error')
df = pd.DataFrame({
    'W_nmos': W_nmos,
    'W_pmos': W_pmos
})
df = df.transpose()
df.columns = ['inv','nor2','nand3','inv']
df * 1e+6
```
